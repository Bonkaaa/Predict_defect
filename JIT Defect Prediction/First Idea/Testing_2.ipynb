```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn.metrics
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.preprocessing import PowerTransformer
from sklearn.metrics import accuracy_score
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import RobustScaler
from sklearn.model_selection import GridSearchCV
```

```python
df_train = pd.read_csv(r'D:\ML\Finding_defect\Files\train.csv')
df_test = pd.read_csv(r'D:\ML\Finding_defect\Files\test.csv')
```

```python

df_train.drop(df_train.columns[0:3], axis =1, inplace= True)
df_train
```

```python
df_test.drop(df_test.columns[0:3], axis =1, inplace= True)
df_test
```

```python
df_train.describe()
```

```python
df_test.describe()
```

```python
y_train = df_train['bug']
y_test = df_test['bug']
y_test
```

```python
df_train.drop(['bug'], axis=1, inplace= True)
df_test.drop(['bug'], axis=1, inplace= True)
X_train = df_train
X_test = df_test
X_test
```

```python
pt = PowerTransformer()
pX_train = pt.fit_transform(X_train)
pX_test = pt.transform(X_test)
```

```python
rs = RobustScaler()
rX_train = rs.fit_transform(pX_train)
rX_test = rs.transform(pX_test)
```

```python
sc = StandardScaler()
zX_train = sc.fit_transform(rX_train)
zX_test = sc.transform(rX_test)
```

```python
zX_train = pd.DataFrame(zX_train)
zX_train.describe()
```

```python
model = LogisticRegression(solver = 'saga', class_weight='balanced', random_state=20, max_iter=10000, penalty= 'l1', C = 1)
model.fit(zX_train, y_train)
```

```python
y_predict = model.predict(zX_test)
y_predict
```
